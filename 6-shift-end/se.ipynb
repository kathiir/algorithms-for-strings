# Shift-And

Первоначальная идея состоит в том, чтобы составить матрицу вида (шаблон == abra, текст == abracadabra):
$$\begin{array}{cc}
   & a & b & r & a & \\
a  & 1 & 0 & 0 & 0 & \\
b  & 0 & 1 & 0 & 0 & \\
r  & 0 & 0 & 1 & 0 & \\
a  & 1 & 0 & 0 & 1 & \\
c  & 0 & 0 & 0 & 0 & \text{// нет в шаблоне -> ноль} \\
a  & 1 & 0 & 0 & 0 & \\
d  & 0 & 0 & 0 & 0 & \text{// нет в шаблоне -> ноль} \\
a  & 1 & 0 & 0 & 0 & \\
b  & 0 & 1 & 0 & 0 & \\
r  & 0 & 0 & 1 & 0 & \\
a  & 1 & 0 & 0 & 1 & \\
\end{array}
$$
Сам алгоритм:
1) Препроцесс - Каждый символ текста представляется в виде одного бита информации (даже если он повторяется. К примеру, маска для aaa: 111), причем, чем ближе буква к началу текста, тем меньше её значение. (для abc: a = 001, b = 010, c = 100).
2) Заводим переменную, которая по разрядности равна длине шаблона.
3) Цикл, на каждом шаге которого мы производим такие действия: (matrix << 1 | 1) & <значение для символа из словаря (пункт 1)>. Сдвиг нужен для того, чтобы знать, что прошлый цикл прошел успешно, а установка 1 в младший бит - чтобы просто заспавнить хоть что-то. Побитовое И нужно для проверки существования ЭТОЙ буквы на ЭТОЙ позиции.
4) Если старший бит оказался равен 1 (мы пишем только в младший), то шаблон совпал успешно.

```python
from shift_end import *

result = shift_and("abra", "abracadabra")
print(result)
```

# Сложность
O(n) - на постоянку для разрядности < разрядности компа.
Ввиду того, что мы обращаемся к каждому символу текста ровно один раз.

Могут быть вопросы про максимальную длину шаблона. Тут интересно, ввиду питоновской реализации.
Для длины шаблона <= разрядности компа (64), сложность O(n)
НО, питон позволяет игнорировать ограничение разрядности (под капотом работает с большей разрядностью).
Тогда для длины шаблона > 64, возможно работать с сложностью O(m*n).
Так как при увеличении длины шаблона в бесконечность, также растет и число операций для осуществления ОДНОГО сдвига.
